## Matplotlib Review

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('classic')
```

```python
import numpy as np

x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

```python
plt.style.use('seaborn-whitegrid')

ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x,np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x,y,c=colors,s=sizes)
plt.colorbar()
plt.title('A Scatterplot Example')
```
